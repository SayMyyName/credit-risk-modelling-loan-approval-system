```python
EXTERNAL_DATASET_PATH = "C:/Users/abhis/Downloads/datasets/External_Cibil_Dataset.xlsx"
INTERNAL_DATASET_PATH = "C:/Users/abhis/Downloads/datasets/Internal_Bank_Dataset.xlsx"
UNSEEN_DATASET_PATH = "C:/Users/abhis/Downloads/datasets/Unseen_Dataset.xlsx"

external_df = pd.read_excel(EXTERNAL_DATASET_PATH)
internal_df = pd.read_excel(INTERNAL_DATASET_PATH)
unseen_df = pd.read_excel(UNSEEN_DATASET_PATH)
```

```python
# Merging the external and internal df based PROSPECTID
df = pd.merge(external_df, internal_df, on="PROSPECTID", how="inner")
```

```python
df.shape
```

```python
df['Approved_Flag'].value_counts(dropna=False)
```

```python
# we'll convert Approved_Flag to binary target
df['target'] = df['Approved_Flag'].map(
    lambda x: 1 if x in ["P1", "P2"] else 0
)

df['target'].value_counts(dropna=False)
```

Target variable is slightly unbalanced, which is normal in credit datasets. 

```python
df.drop(columns="Approved_Flag", inplace=True)
```
